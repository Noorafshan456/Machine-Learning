# Filling Null values in data_train

#### Filling Null values in data_train

Defining missing_value_table it will return the number of missing value and the percentage of missing values in decending order

```python
def missing_values_table(df):
    # Check the total number of missing values
    mis_val=df.isnull().sum()    
    # Check the proportion of missing values
    mis_val_percent=100*df.isnull().sum()/len(df)    
   # Create a table of missing values and weight of missing values
    mis_val_table=pd.concat([mis_val, mis_val_percent],axis=1)    
   # Rename column
    mis_val_table_ren_columns=mis_val_table.rename(columns={0:'Missing Values',1:'% of Total Values'})
    # Sort except for columns with missing values of 0
    mis_val_table_ren_columns=mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:,1]!=0].sort_values('% of Total Values',ascending=False).round(1)
    #Print summary results
    print("The total number of columns in train is "+str(df.shape[1])+
         " Among them, the number of columns with missing values is "+ str(mis_val_table_ren_columns.shape[0]))   
    return mis_val_table_ren_columns
```

#### defining the function to adjust X_test columns

```python
def adjust_columns(data_train, data_test):
    train_columns = []
    test_columns = []
    
    # Get data_train columns not present in data_test
    for column in data_train.columns:
        if column not in data_test.columns:
            train_columns.append(column)
    
    # Get data_test columns not present in data_train
    for column in data_test.columns:
        if column not in data_train.columns:
            test_columns.append(column)
    
    # Remove test_columns from X_test
    data_test = data_test.copy()
    data_test = data_test.drop(test_columns, axis=1)
    
    # Add train_columns and fill with 0's
    for column in train_columns:
        column_index = list(data_train.columns).index(column)
        data_test.insert(loc=column_index, column=column, value=0)
    
    return data_test
```
